```python
gender = survey_2016_pt7["What is your gender?"].unique()
```

```python
survey_2016_pt7_renamed["Gender"] = survey_2016_pt7["What is your gender?"].astype(str)
print(gender)
```

```python
#Made gender groups
male_str = ['MALE','Male','Male ','male ', 'male', 'Male','M','m','man','Male.', 'male 9:1 female, roughly','Man', 'Sex is male','Malr',"I'm a man why didn't you make this a drop down question. You should of asked sex? And I would of answered yes please. Seriously how much text can this take? ", 'mail', 'M|', 'Male/genderqueer','male'] 
 
trans_str = ['Bigender', 'non-binary','Cis female ', 'Transitioned, M2F', 'Other/Transfeminine', 'Cis male', 'Androgynous','Male (cis)', 'nb masculine', 'Cisgender Female','cis male', 'Agender','Unicorn', 'Cis Male', 'Male (trans, FtM)', 'Cis-woman','cisdude', 'Genderflux demi-girl','cis man', 'AFAB','Transgender woman']          
female_str = ['Female','female','I identify as female.', 'female ','Female assigned at birth ', 'F', 'Woman','fm','f','Genderfluid (born female)','Female or Multi-Gender Femme', 'Female ','woman', 'female/woman','genderqueer woman', 'Female (props for making this a freeform field, though)', ' Female','female-bodied; no feelings about gender']
other_str =['nan', 'Other', 'none of your business',
       'genderqueer', 'Human', 'Genderfluid', 'Enby', 'mtf', 'Queer',
       'Dude', 'Fluid', 'fem', 'Nonbinary', 'human', 'Genderqueer']
for (row, col) in survey_2016_pt7_renamed.iterrows():

    if col.Gender in male_str:
        survey_2016_pt7_renamed['Gender'].replace(to_replace=col.Gender, value='male', inplace=True)

    if col.Gender in female_str:
        survey_2016_pt7_renamed['Gender'].replace(to_replace=col.Gender, value='female', inplace=True)

    if col.Gender in trans_str:
        survey_2016_pt7_renamed['Gender'].replace(to_replace=col.Gender, value='trans', inplace=True)
    if col.Gender in other_str:
        survey_2016_pt7_renamed['Gender'].replace(to_replace=col.Gender, value='others', inplace=True)
```
