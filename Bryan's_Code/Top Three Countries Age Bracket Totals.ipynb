```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Updated_2017 = "reduced_updated_2017.csv"

updated_17_pd = pd.read_csv(Updated_2017)
updated_17_pd.head()
```

```python
#updated_17_pd.plot.bar(x='age_bins',y='Age_Int')
```

```python
filter_country_age = updated_17_pd.filter(["Country", "Age_Int"]) 
filter_country_age
```

```python
filter_after = filter_country_age[(filter_country_age['Country'] == "United States of America") | (filter_country_age['Country'] == "United Kingdom") | (filter_country_age['Country'] == "Canada")]
filter_after
```

```python
country_count = updated_17_pd['Country'].value_counts()
country_count
```

```python
top_three_countries = country_count.loc[['United States of America','United Kingdom', 'Canada']]
top_three_countries
```

```python
filter_after.groupby(['Age_Int','Country']).size().unstack().plot(kind='bar',stacked=True)
#plt.xlim(-0.25)
#plt.ylim(0, max(Country)+0.4
plt.title("Top Three Countries Age Bracket Totals")
plt.ylim(-0.25)
plt.xlabel("Age Brackets", fontsize=10)
plt.ylabel("Totals", fontsize=10)
plt.tight_layout()
plt.grid()
plt.show()
plt.savefig("Images/2017_Age_Totals_By_Top_Three_Countries.png")

```
