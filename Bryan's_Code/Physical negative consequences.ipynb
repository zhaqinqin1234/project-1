```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_2017 = "survey_2017.csv"
df_17 = pd.read_csv(data_2017)

data_2016 = "survey_2016.csv"
df_16 = pd.read_csv(data_2016)

data_2014 = "survey_2014.csv"
df_14 = pd.read_csv(data_2014)

data_2018 = "survey_2018.csv"
df_18 = pd.read_csv(data_2018)

data_2019 = "survey_2019.csv"
df_19 = pd.read_csv(data_2019)
```

```python
negative_2014 = df_14["Would you bring up a physical health issue with a potential employer in an interview?"]
negative_2016 = df_16["Do you think that discussing a physical health issue with your employer would have negative consequences?"]
```

```python
negative_2014.value_counts()
```

```python
negative_2016.value_counts()
```

```python
counts_2014 = negative_2014.value_counts()
counts_2016 = negative_2016.value_counts()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(counts_2014.index, counts_2014.values, alpha=0.8)
plt.title(' 2014 Discussing a physical health issue with your employer would have negative consequences?')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Answers', fontsize=12)
plt.grid()
plt.show()
plt.savefig("Images/2014_Answers_Physical_Negative consequences.png")
```

```python
plt.figure(figsize=(10,5))
sns.barplot(counts_2016.index, counts_2016.values, alpha=0.8)
plt.title(' 2016 Discussing a physical health issue with your employer would have negative consequences?')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Answers', fontsize=12)
plt.grid()
plt.show()
plt.savefig("Images/2016_Answers_Physical_Negative consequences.png")
```
